```
import pymysql


def get_input():
    flag = False

    while flag == False:
        print('Input the Number to Perform that Action.')
        print('(1) Add a Student')
        print('(2) Delete a Student based on their student ID')
        print('(3) modify a students name based on their student ID')
        print('(4) Get the total count of students in one specific major')
        print('(5) Get one students ID and name Randomly')
        print('(6) Exit the program')
        user_input = int(input('Number: '))

        if user_input == 1:
            insert_student()
            input('Press any button to continue......')
        elif user_input == 2:
            delete_student()
            input('Press any button to continue......')
        elif user_input == 3:
            modify_student_name()
            input('Press any button to continue......')
        elif user_input == 4:
            total_in_major()
            input('Press any button to continue......')
        elif user_input == 5:
            get_random_student()
            input('Press any button to continue......')
        elif user_input == 6:
            flag = True
            print('Goodbye!')
        else:
            print('That was not a valid input!')
            input('Press any button to continue......')


def insert_student():
    stuid = input('What is the students ID number?')
    stuname = input('What is the students name?')
    major = input('What is the students Major?')
    stuyear = input('What is the students graduation year?')
    tuition = input('What is the students tuition"')
    age = input('What is the students age?')

    db = pymysql.connect("localhost","root","1234","pydb")
    dbcursor = db.cursor()

    sql="insert into tstudents(stuid,stuname,major,stuyear,tuition,age) values(%s,%s,%s,%s,%s,%s)"
    values=(stuid,stuname,major,stuyear,tuition,age)

    dbcursor.execute(sql,values)
    db.commit()

    dbcursor.close()
    db.close()


def delete_student():
    db = pymysql.connect("localhost", "root", "1234", "pydb")
    dbcursor = db.cursor()

    sql = "delete from tstudents where stuid=%s"
    values = int(input("please input an id you want to delete:"))

    dbcursor.execute(sql, values)
    db.commit()

    dbcursor.close()
    db.close()


def modify_student_name():
    stuid = input('What is the ID of the student\'s name you would like to replace? ')
    stuname = input('What is the students new name? ')


    db = pymysql.connect("localhost", "root", "1234", "pydb")
    dbcursor = db.cursor()

    sql = "update tstudents set stuname=%s where stuid=%s"
    value = (stuname,stuid)

    dbcursor.execute(sql,value)
    db.commit()

    dbcursor.close()
    db.close()


def total_in_major():
    major = input('What is the major that you want to search for? ')

    db = pymysql.connect("localhost", "root", "1234", "pydb")
    dbcursor = db.cursor()

    sql = "select * from tstudents where major like %s"
    value = major

    dbcursor.execute(sql, value)
    results = dbcursor.fetchall()

    dbcursor.close()
    db.close()

    print(len(results))


def get_random_student():
    db = pymysql.connect("localhost", "root", "1234", "pydb")
    dbcursor = db.cursor()

    sql = "select * from tstudents order by rand() limit 1"

    dbcursor.execute(sql)
    results = dbcursor.fetchall()

    dbcursor.close()
    db.close()

    print("Student ID \t Student Name")
    print("-----------------------")
    for row in results:
        print(str(row[0])+"\t"+"\t"+"\t"+str(row[1]))
        print("-----------------------")


get_input()
```
