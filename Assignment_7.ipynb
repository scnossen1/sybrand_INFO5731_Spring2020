```
import turtle
import time



def create_window():
    window = turtle.Screen()
    window.bgcolor('green')
    window.title('Turtle')
    return window


def create_turtle(name):
    name = turtle.Turtle()
    name.pensize(3)
    return name


def draw_squares(name, size, count):
    for i in range(4):
        name.forward(size)
        name.left(90)

    count -= 1

    if count > 0:
        name.penup()
        name.forward(size * 2)
        name.pendown()
        draw_squares(name, size, count)

    name.reset()


animal = str(input('What do you want the turtles name to be? '))
count = int(input('How many squares do you want there to be? '))
size = int(input('How Big do you want the squares to be? '))
create_window()
animal_name = create_turtle(animal)
draw_squares(animal_name, size, count)



def draw_boxes(name, number, size):
    for i in range(4):
        name.forward(size)
        name.left(90)

    name.penup()
    name.right(90)
    name.forward(5)
    name.right(90)
    name.forward(5)
    name.left(180)
    name.pendown()
    size += 10
    number += -1

    if number > 0:
        draw_boxes(name, number, size)



create_window()
draw_boxes(animal_name, count, size)



int_list = [5, 15, [23, 5, [66, 17, 5,[54, 80, 5,[111, 5,[154, 5,]]], 87, 790,]], 15, 5]

def count_total(nested_list, number):
    count = 0
    for element in nested_list:
        if type(element) is list:
            if element == number:
                count += 1
            count += count_total(element, number)
        else:
            if element == number:
                count += 1
    return count


print(count_total(int_list, 5))
```
