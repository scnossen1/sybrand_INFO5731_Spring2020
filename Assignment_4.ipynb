```
import turtle
import time
import numpy as np


print('\n')
print('Part 1 and 3')
print('\n')

def create_window():
    window = turtle.Screen()
    window.bgcolor('green')
    window.title('Turtle')
    return window


def create_turtle(name):
    name = turtle.Turtle()
    name.pensize(3)
    return name


def draw_square(name, size):
    for i in range(4):
        name.forward(size)
        name.left(90)

    size += 10


def draw_boxes(name):
    number = int(input('How many boxes? '))
    size = 10
    for i in range(number):
        draw_square(name, size)
        name.penup()
        name.right(90)
        name.forward(5)
        name.right(90)
        name.forward(5)
        name.left(180)
        name.pendown()
        size += 10

    time.sleep(2)
    name.reset()


def draw_maze(name):
    length = 1
    lines = int(input('How many lines? '))
    for i in range(lines):
        name.forward(length)
        name.left(90)
        length += 1
    time.sleep(2)
    name.reset()


animal = input('What is this animals name? ')
window_complete = create_window()
animal_complete = create_turtle(animal)
draw_boxes(animal_complete)
draw_maze(animal_complete)


print('\n')
print('Part 3')
print('\n')

def extension(string):
    dot_location = string.rfind('.')
    if dot_location == -1:
        print('This is not a valid input!')
    else:
        new_string = string[(dot_location + 1):]
        print('Your extention type is: ' + new_string)

filename = str(input('Give me a filename and its Extension: '))
extension(filename)


print('\n')
print('Part 4')
print('\n')

sentence = list('I love cats! I drive lots of cats! Can I borrow your cat?')

def replace(string, old, new):
    new_string = np.chararray(len(string))
    for i in range(len(string)):
        if string[i] == old:
            string[i] = new
    new_string = ''.join(string) 
    print(new_string)

print(replace(sentence, 't', 'r'))


```
