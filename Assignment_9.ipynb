```
import math


def print_dif(difference):
    if (difference > 0):
        print('Time 1 is ', abs(difference), ' minutes larger than Time 2!')
    elif(difference < 0):
        print('Time 2 is ', abs(difference), ' minutes larger than Time 1!')
    else:
        print('Time 1 and Time 2 are the same!')


class Mytime:
    def __init__(self, hrs=0, mins=0, secs=0):
        totalsecs = hrs * 3600 + mins * 60 + secs
        self.hours = totalsecs // 3600
        leftoversecs = totalsecs % 3600
        self.minutes = leftoversecs // 60
        self.seconds = leftoversecs % 60

    def increment(self, seconds):
        self.seconds += seconds;
        while self.seconds >= 60:
            self.seconds -= 60
            self.minutes += 1

        while self.minutes >= 60:
            self.minutes -= 60
            self.hours += 1

    def to_seconds(self):
        return self.hours * 3600 + self.minutes * 60 + self.seconds

    def compare(self, time):
        dif_hr = (self.hours - time.hours) * 60
        dif_min = self.minutes - time.minutes
        dif_sec = round((self.seconds - time.seconds) / 60)

        total_dif = dif_hr + dif_min + dif_sec
        return total_dif

    def __add__(self, other):
        return Mytime(0, 0, self.to_seconds() + other.to_seconds())

    def __str__(self):
        return "the time is {0}:{1}:{2}".format(self.hours, self.minutes, self.seconds)


def add_time(t1, t2):
    h = t1.hours + t2.hours
    m = t1.minutes + t2.minutes
    s = t1.seconds + t2.seconds

    if s >= 60:
        s -= 60
        m += 1

    if m >= 60:
        m -= 60
        h += 1

    sum_t = Mytime(h, m, s)
    return sum_t


time1 = Mytime(11, 10, 0)
time2 = Mytime(12, 0, 31)
print_dif(time1.compare(time2))



class Card:
    suits = ["Clubs", "Diamonds", "Hearts", "Spades"]
    ranks = ["narf", "Ace", "2", "3", "4", "5", "6", "7", "8", "9", "10", "Jack", "Queen", "King"]

    def __init__(self, suit=0, rank=0):
        self.suit = suit
        self.rank = rank

    def __str__(self):
        return (self.ranks[self.rank] + " of " + self.suits[self.suit])

    def cmp(self, other):
        if self.suit > other.suit: return 1
        if self.suit < other.suit: return -1

        if self.rank == 1 and other.rank == 13: return 1
        if self.rank == 13 and other.rank == 1: return -1

        if self.rank > other.rank: return 1
        if self.rank < other.rank: return -1

        return 0

    def __eq__(self, other):
        return self.cmp(other) == 0

    def __le__(self, other):
        return self.cmp(other) <= 0

    def __ge__(self, other):
        return self.cmp(other) >= 0

    def __gt__(self, other):
        return self.cmp(other) > 0

    def __lt__(self, other):
        return self.cmp(other) < 0

    def __ne__(self, other):
        return self.cmp(other) != 0


three_of_clubs = Card(0, 3)

card1 = Card(1, 1)
print(card1)

card2 = Card(1, 13)
print(card2)

print(card1 > card2)



def read_pos_int():
    try:
        number = int(input("Enter a positive number: "))
    except ValueError:
        print('That is not a valid number!')
    else:
        if number > 0:
            print('Youur number is', number, 'and it is positive!')
        elif number == 0:
            print('0 is neither a positive nor negative number!')
        else:
            print('That is a negative number, and I asked for a positive number!')



read_pos_int()
```
